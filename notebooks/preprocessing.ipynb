```python
import pandas as pd
import os
```

```python
base_url = '../data/raw/2019/'

# getting files / folders from directory of 2019'th selection
last_data = os.listdir(base_url)

# getting csv files
csv_files = [file for file in last_data if '.csv' in file]

# getting folders
folders = [file for file in last_data if '.csv' not in file]
```

```python
# creating dataframe for scoring all data
final_df = pd.DataFrame()


# creating by_regions folder to store data
if 'by_regions' not in os.listdir(base_url):
    os.makedirs(f'{base_url}by_regions')


# iterating through csv_files list
for file_name in csv_files:
    # folder_name.csv = file_name   →   folder_name = file_name.replace('.csv', '')
    folder_name = file_name.replace('.csv', '')
    
    # getting files from the folder
    files_in_folder = os.listdir(f'{base_url}{folder_name}')

    # creating df to store data for each region for candidate
    df_candidate = pd.DataFrame()

    # iterating through files_in_folder list
    for folder_file in files_in_folder:
        # setting file path
        file_path = f'{base_url}{folder_name}/{folder_file}'
        
        # reading file to df
        df = pd.read_csv(file_path)
        
        # creating region column
        df['Область'] = ' '.join(folder_file.replace('.csv', '').split('_'))

        # concating with final dataframe
        df_candidate = pd.concat([df_candidate, df], axis=0)

        # saving to file
        df_candidate.to_csv(f'{base_url}by_regions/{folder_name}_by_regions.csv', index=False)


    # setting file path
    file_path = f'{base_url}{file_name}'

    # reading csv file
    df = pd.read_csv(file_path)

    # creating candidate column
    df['Кандидат'] = ' '.join(folder_name.split('_'))

    # concating with final dataframe
    final_df = pd.concat([final_df, df], axis=0)


# saving to file
final_df.to_csv(f'{base_url}final_results_by_regions.csv', index=False)
```

```python
# link to regions folder
link = '../data/raw/2019/by_regions'

# getting all files from directory
all_files = os.listdir(link)

# creating dataframe for storing all data
df_regions_districts = pd.DataFrame()


# iterating through them
for file in all_files:
    # getting name of candidate
    name = file.replace('_by_regions.csv', '').replace('_', ' ')
    
    # full link to file
    full_link = f'{link}/{file}'

    # reading file to csv
    df = pd.read_csv(full_link)

    # creating candidate column
    df['Кандидат'] = name

    # concating with final dataset
    df_regions_districts = pd.concat([df_regions_districts, df], axis=0)


# creating final folder
os.makedirs('../data/raw/2019/final')

# saving dataframe to file
df_regions_districts.to_csv(f'../data/raw/2019/final/candidate_district.csv', index=False)
```
